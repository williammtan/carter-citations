```python
from scholarly import scholarly, ProxyGenerator
from scraperapi_sdk import ScraperAPIClient
import pandas as pd

SCRAPER_API_KEY = 'API_KEY'

pg = ProxyGenerator()
success = pg.ScraperAPI(SCRAPER_API_KEY)

scraper_client = ScraperAPIClient(SCRAPER_API_KEY)

scholarly.use_proxy(pg)
```

```python
# Retrieve the publication
search_query = scholarly.search_pubs("Correlation of Structural and Electronic ")
pub = next(search_query)
```

```python
import os
import PyPDF2
import requests
from openai import OpenAI
from io import BytesIO
import html2text

# Initialize OpenAI client
client = OpenAI(api_key="API_KEY")

def extract_text_from_pdf_bytes(pdf_bytes):
    """
    Extracts text from a PDF byte string.

    Args:
        pdf_bytes (bytes): The content of the PDF as a byte string.

    Returns:
        str: Extracted text from the PDF.
    """
    try:
        # Use BytesIO to handle the byte string as a file-like object
        with BytesIO(pdf_bytes) as pdf_stream:
            reader = PyPDF2.PdfReader(pdf_stream)
            text = ""
            for page in reader.pages:
                page_text = page.extract_text()
                if page_text:
                    text += page_text
        return text.strip()
    except Exception as e:
        print(f"Error reading PDF content: {e}")
        return ""

# Function to download PDF from URL
def download_pdf(url):
    try:
        text = scraper_client.get(url=url)
        return text
    except Exception as e:
        print(f"Error downloading {url}: {e}")
        return None


def extract_abstract(paper_text):
    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": f"Extract this paper's abstract and don't return anything else. Do not respond if you don't know the answer or the answer is not given. Paper: {paper_text[:104000]}"}
            ]
        )
        # Extract the assistant's reply
        summary = response.choices[0].message.content.strip()
        return summary
    except Exception as e:
        print(f"Error generating summary: {e}")
        return None

def extract_summary(paper_text):
    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": f"TLDR this paper and don't return anything else (not even paper title and authors). Do not respond if you don't know the answer or the answer is not given. Paper: {paper_text[:104000]}"}
            ]
        )
        # Extract the assistant's reply
        summary = response.choices[0].message.content.strip()
        return summary
    except Exception as e:
        print(f"Error generating summary: {e}")
        return None

# Iterate over a list of PDF URLs
def summarize_paper_from_url(url):
      print(f"Processing {url}...")
      data = download_pdf(url)
      if data is None:
        return {
            'abstract': None,
              'summary': None
        }

      data = html2text.html2text(data)
      # if url.endswith('.pdf'):
      #   data = extract_text_from_pdf_bytes(data)

      if data:
          abstract = extract_abstract(data)
          # summary = extract_summary(data)
          return {
              'abstract': abstract,
              # 'summary': summary
          }

      else:
          return None
```

```python
for i, row in df.iterrows():
  if not pd.isna(row['pub_url']):
    r = summarize_paper_from_url(row['pub_url'])
    print(r)
    if r != None:
      # df.loc[i, 'summary'] = r['summary']
      df.loc[i, 'abstract'] = r['abstract']
    else:
      df.loc[i, 'abstract'] = row['bib']['abstract']
      # df.loc[i, 'summary'] = 'N/A'
  else:
    df.loc[i, 'abstract'] = row['bib']['abstract']
    # df.loc[i, 'summary'] = 'N/A'
```

```python
# RANKED BY # CITATIONS


ranked_citations_summary = ""
for i, c in enumerate(df.sort_values(by="num_citations", ascending=False).iterrows()):
  c = c[1]
  ranked_citations_summary += f"Paper {i}: [{c['bib']['title']}]({c['pub_url']}) ({c['bib']['pub_year']}, {c['num_citations']} cites)\n Abstract: {c['abstract']}\n TL;DR:{c['summary']} \n" + "\n"

with open("citations.md", "w") as f:
  f.write(ranked_citations_summary)
```
