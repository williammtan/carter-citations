```python
from scholarly import scholarly, ProxyGenerator
import pandas as pd

pg = ProxyGenerator()
success = pg.ScraperAPI("API_KEY")

scholarly.use_proxy(pg)
```

```python
# Retrieve the publication
search_query = scholarly.search_pubs("Correlation of Structural and Electronic ")
pub = next(search_query)
```

```python
pub
```

```python
citations = list(scholarly.citedby(pub))
df = pd.DataFrame(citations)
```

```python
print([citation['bib']['abstract'] for citation in citations])
```

```python
citations[0]
```

```python
df
```

```python
# RANKED BY # CITATIONS

ranked_citations_summary = ""
for i, c in enumerate(df.sort_values(by="num_citations", ascending=False).iterrows()):
  c = c[1]
  ranked_citations_summary += f"Paper {i}: [{c['bib']['title']}]({c['pub_url']}) ({c['bib']['pub_year']}, {c['num_citations']} cites)\nAbstract: {c['bib']['abstract']}\n" + "\n"

with open("citations.md", "w") as f:
  f.write(ranked_citations_summary)
```
